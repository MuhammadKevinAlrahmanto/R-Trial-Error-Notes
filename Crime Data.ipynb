```r
library(readxl)
```

```r
crime_data=read_xls('FBI Crime Data.xls')
crime_data
```

```r
crime_data[3,]
```

```r
colnames(crime_data) = crime_data[3,]
crime_data
```

```r
crime_data['converted_index']=(
    sapply(crime_data[,1], as.numeric)
)
```

```r
crime_data
```

```r
crime_data_filter=(
crime_data[!is.na(crime_data['converted_index']),]
)
crime_data_filter
```

```r
years = substr(crime_data_filter$Year, 1, 4)
crime_data_filter$Year=years
crime_data_filter
```

```r
# Drop Columns
remove_col=c('Year', 'converted_index')
crime_data_filter=crime_data_filter[,!colnames(crime_data_filter) %in% remove_col]
```

```r
summary(crime_data_filter)
```

```r
crime_data_filter_n=sapply(crime_data_filter, as.numeric)
crime_data_filter_n
```

```r
summary(crime_data_filter_n)
```

```r
years
```

```r
rownames(crime_data_filter_n)=years
```

```r
rownames(crime_data_filter_n)
```

```r
crime_data_filter_n['2012',]
```

```r
grepl('rate', colnames(crime_data_filter_n))
```

```r
rate_crime_data=(
    crime_data_filter_n[,grepl('rate', colnames(crime_data_filter_n))]
)
rate_crime_data
```

```r
nominal_crime_data=rate_crime_data=(
    crime_data_filter_n[,!grepl('rate', colnames(crime_data_filter_n))]
)
nominal_crime_data
```

```r
colnames(nominal_crime_data)<- c('pop', 'vio cri', 'mur', ' robb', 'agg assult','pro crime', 'burg', 'theft', 'mot theft') 
```

```r
nominal_crime_data
```

```r
class(nominal_crime_data)
```

```r
means_crime=as.data.frame(sapply(as.data.frame(nominal_crime_data), mean))
colnames(means_crime)='mean_value'
means_crime
```

```r
means_crime[2:nrow(means_crime), ]
```

```r
means_crime$mean_value %in% means_crime[2:nrow(means_crime), ]
```

```r
means_crime=tail(means_crime,-1)
means_crime
```

```r
library(ggplot2)
```

```r
ggplot(data=means_crime,
      aes(x=rownames(means_crime),
         y=mean_value)) + geom_bar(stat='identity', fill='blue')
```
