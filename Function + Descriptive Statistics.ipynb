```r
count_vowel<- function(text){
    split_text=strsplit(tolower(text), '')[[1]]
    count=0
    for (letter in split_text){
        if (letter %in% c('a', 'i', 'u', 'e', 'o') ){
            count=count+1
        }
    } 
    return(count)
}
count_vowel('fOuNd A vOwEL')
```

```r
#Control FLow 
custom_mean_for <- function(my_vector){
    sum=0
    for (element in my_vector){
        sum=sum + element
    }
    mean_vector=sum/length(my_vector)
    return (mean_vector)
}
custom_mean_for(c(10,20,30))
```

```r
# If statements
custom_mean_if <- function(my_vector){
    
    if (class(my_vector)=='character'){
        print('Numeric vector please!')
    }
    else{
        sum=0
        for (element in my_vector){
            sum=sum + element
        }
        mean_vector=sum/length(my_vector)
        return (mean_vector)
        }
}
custom_mean_if(c(10,'0',30))
```

```r
# While loop
factorial <- function(num){
    if(num == 0 ){return(1)}
    else if (num <0){return ('Positive number please')}
    else{
        my_vector=1:num
        result=1
        i=1
        while (i <= length(my_vector)){
            result=result*my_vector[i]
            i=i+1
        }
        return(result)
        }
}
factorial(4)
```

```r
library(ggplot2)
```

```r
plot_store_department <- function(store_n, dept_n){
    walmart_filter = walmart_data[
        (walmart_data$Store == store_n) & (walmart_data$Dept == dept_n), ]
    ggplot(
        data = walmart_filter,
        aes(x=as.Date(Date), y=Weekly_Sales, group=1)
    ) + geom_line(color='darkorange') + geom_point (color='orange')
}
```

```r
plot_store_department(1,90)
```

```r
plot_group_department <- function(store_n, dept_n, colorarg){
    walmart_filter = walmart_data[
        (walmart_data$Store == store_n) & (walmart_data$Dept %in% dept_n), ]
    ggplot(
        data = walmart_filter,
        aes(x=as.Date(Date), y=Weekly_Sales, group=Dept, color=as.factor(Dept))
    ) + geom_line() + geom_point ()
}
```

```r
plot_group_department(10, c(1,2,3,4))
```
