#### Group

For most applications the grouping is set implicitly by mapping one or more discrete variables to x, y, colour, fill, alpha, shape, size, and/or linetype. That means that every time that you are defining a color, fill or something like this, you are implicitly defining a group.

So most of the time you won't have to worry about the group aesthetic. But, the thing is that the group aesthetic is by default set to the interaction of all discrete variables in the plot. This is sometimes what you want, sometimes it is not. Take for example this dataset, that we also use on the course, the Oxboys one. Here we have different observations of different subjects, and if we plot it without grouping, the result is not what we want:

```r
Oxboys<-nlme::Oxboys
```

```r
head(Oxboys) 
```

```r
unique(Oxboys$Subject)
```

```r
ggplot(xboys, aes(age, height)) + geom_line()
```

```r
ggplot(Oxboys, aes(age, height)) + geom_line(aes(color=Subject))
```

So this works, now I have a different line for every Subject. But let's say that I don't want color on this one. How can I specify the grouping without color? I can do it by specifying the group aesthetic:

```r
ggplot(nlme::Oxboys, aes(age, height)) + geom_line(aes(group=Subject))
```

So basically, the group aesthetic is something you don't have to worry about unless your plot looks weird. Most of the times, the default grouping works well, and, because we normally use aesthetics such as fill or color, the grouping generated by these aesthetics is enough. Only when it does not work, then we have to "force" the grouping into what we want.

#### Scales 
- Scales control the mapping from data to aesthetics. They take your data and turn it into something that you can see, like size, color, position or shape. They also provide the tools that let you interpret the plot: the azes and lege
- Domain of data => Scales => Domain of scales
- Domain of aesthetis => Guides(axes and legends) => Domain of data
- In short: Sclaes generate gudes. You can't see scales, but you can see guides.
- Axis(axis label, tick mark and label)
- Legend(legend titile, key, key label)
