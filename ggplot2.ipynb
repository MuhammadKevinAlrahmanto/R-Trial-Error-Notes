```r
library(ggplot2)
```

```r
A = c(2, 1, 4, 9)
B = c(3, 2, 5, 12)
C = c(4, 1, 15, 8)
D = c('a', 'a', 'b', 'b')
my_data <- data.frame(A, B, C, D)
my_data
```

```r
p <- ggplot(data=my_data, mapping = aes(x=A, y=C)) 
p
```

```r
p<- p + geom_point(mapping=aes(shape=D), stat="identity", size=4, position='identity') 
p
```

```r
p <- p + scale_x_continuous() + scale_y_log10()
p
```

```r
p <- p+ coord_polar()
p
```

```r
p <- p + coord_cartesian()
p
```

```r
p <- p + facet_grid(cols = vars(D)) 
p
```

```r
p2 <- ggplot(my_data, aes(x=A, y=C)) + geom_point(aes(shape=D), size=4) + scale_y_log10() + facet_grid(cols = vars(D)) 
p2
```
